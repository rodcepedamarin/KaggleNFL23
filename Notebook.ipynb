# Eye on data: Defensive Linemen Field of Vision Occupancy - DELFOS. 

## Introduction

One of the most direct ways to measure defensive linemen is the ability to get to the quarterback and ultimately cause a sack or fumble, but when the quarterback passes the football, is there any way to measure how defensive linemen affected the outcome of the pass, whether this is a complete pass or an incomplete pass?

This question is taken into study in this project, where <u>the impact of the defensive linemen on pass plays</u> is measured, and for the matter, the **Defensive Linemen Field Of View Occupancy** metric (DELFOS from now on) is presented. DELFOS measures the percentage of visual space that defensive linemen occupy in the quarterback's field of vision at the precise moment of the pass.

The height and width of the defensive linemen, the distance to the quarterback in the moment of the pass, and the ability to drag offensive linemen players into the passer position are factors that will condition the percentage of visual field occupied and will potentially interfere with the quarterback's ability to pass in clear view conditions.

## DELFOS Calculation Methodology

**To build DELFOS, first, the point of vision (POV) of the quarterback in the exact moment of the pass needs to be replicated, calculating then the percentage of visual space occupied by the defensive linemen in the football field withing visual reach to the quarterback.** To replicate a field of vision, the following assumptions have been taken:
1. The width of vision of all players is 120º
2. The height of vision of all players is 90º
3. The distance from the top of the head to the eyeball is 7cm (2.75 in)
4. The body width is directly proportional with the player weight
5. **The field of vision of the quarterback is represented by a 300x519 matrix** where $y_{fv} = 300$ and $x_{fv} = 519$.

The two dimensional position of the defensive players in that matrix will be determined by:
* The Y-axis position
* The X-axis position

### The Y-axis position

The vertical position of the players in the quarteback's field of vision will be parametrized with two measurements: the position of the feet $y_g$ and the position of the head $y_h$.

<img src="img/image12.png" width="100%" />

The total field of vision for the distance at which the defender is located is $h_{fv} = 2d·tan(60)$.

The position of the feet in the quarterback's field of vision is given by
$$
y_{g} 
= 
\begin{cases}
y_{fv}\frac{(d-d_1)tan(60)}{2d·tan(60)} \text{ if } d_1 < d\\
0 \text{ if } d_1 > d\\
0 \text{ if } d_1 = d\\
\end{cases}
$$

The position of the head in the quarterback's field of vision is given by
$$
y_{h} 
= 
\begin{cases}
y_g + y_{fv}\frac{h}{2d·tan(60)} \text{ if } d_1 < d\\
y_{fv}\frac{h-((d_1-d)tan(60))}{2d·tan(60)} \text{ if } d_1 > d\\
y_{fv}\frac{h}{2d·tan(60)} \text{ if } d_1 = d\\
\end{cases}
$$


### The X-axis position

The horizontal position of the defensive player in the quarteback's field of vision will be parametrized with three measurements: the position of the center of the player $x_0$, the leftmost part of the player $x_0 - \frac{w}{2}$ and the rightmost part of the player $x_0 + \frac{w}{2}$. 

<img src="img/image7.png" width="100%" />

The $x_r$ and $x_l$ points are given by the intersection of $l$ and the lines originated at $x_{qb}$ with the  angles $\omega \pm \frac{120}{2}$, where $\omega$ is the orientation angle, $r$ is the line that passes through the point $x_{qb}$ and has an angle $\omega$, and $l$ is the perpendicular line to $r$ passing through $x_{def}$.

$$x_0 = x_{fv}\frac{d(x_l, x_{def})}{d(x_l, x_r)}$$
$$x_{0_{-}} = x_0 - \frac{\omega}{2}$$
$$x_{0_{+}} = x_0 + \frac{\omega}{2}$$

### DELFOS Construction

Lastly, for constructiong DELFOS, ie, constructing the metric that measures the percentual field of vision of the QB occupied by the defensive linemen, two things need to be done: 

1. Finding the offensive players are located in the field conditioned by the presence of the defensive linemen. For this, a simple Nearest Neighbour (NN = 1) method will be used.
2. Removing all the players that don't belong to the defensive line nor are influenced by it.

### Future Work

In this analysis, measuring the impact of the defensive linemen has been the main focus, but for future work, the idea behind the metric could be of use: other sports like soccer are developing metrics and measurements based on the position in the field of the players. The new VAR Offside introduced at The Qatar World Cup, as well as the Goal Probability are two examples. This technology calculates on real time the 3 dimmensional position of the players in the field and performs the necessary calculations.

<img src="img/image16.png" width="100%" />

Something simmilar could be developed in the NFL in real time for television purposes, where a more complex engine can calculate the visual field of the quarterback after the pass and estimate the probability of completion based of the available field of vision.
